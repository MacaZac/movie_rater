```python
def create_view_from_csv(file_name, file_mask = '*.tsv.gz'):
    file_path = 'file:///home/jovyan/work/not_work/movie_rater/data/input'
    df = spark.read.csv(f'{file_path}/{file_name}{file_mask}',sep=r'\t', header= True)
    df.createOrReplaceTempView(file_name)
    return df
```
